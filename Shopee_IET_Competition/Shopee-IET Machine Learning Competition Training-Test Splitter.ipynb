# Shopee-IET Machine Learning Competition Training-Test Splitter

### 1. import modules

```python
import os
import shutil
from distutils.dir_util import copy_tree
import random
import pandas as pd
```

### 1. Clone the raw data into a destination folder

```python
def copy_folder(src, dest): 
    
    if not os.path.exists(dest):
        os.makedirs(dest)
    
    
    copy_tree(src, dest)
```

```python
source_folder = 'F:\Data\Demo Data-Shopee-IET'
dest_folder = 'F:\Data\Splitted Demo Data-Shopee-IET'
```

```python
copy_folder(source_folder, dest_folder)
```

### 2. Create a test folder in the destination folder

```python
test_folder = dest_folder + "\Test"

if not os.path.exists(test_folder):
        os.makedirs(test_folder)
```

### 3. Move 30% of raw data into the test_folder

```python
folder_full_names = [os.path.join(dest_folder, folder) for folder in os.listdir(dest_folder) if not folder == "Test"]
```

```python
folder_full_names
```

```python
for folder_full_name in folder_full_names:
    file_names = os.listdir(folder_full_name)
    i = 1
    j = 1
    for file_name in file_names:
        if i > 70:
            file_full_name = os.path.join(folder_full_name, file_name)
            folder_name = folder_full_name.split("\\",8)[-1]
            shutil.move(file_full_name, test_folder + '/' + "Test_" + folder_name + "_" + str(j) + ".jpg")
            j = j + 1
            
        i = i + 1
        
        
        if i == 101:
            i = 1
```

### 4. Rename all the pictures

```python
for folder_full_name in folder_full_names:
    file_names = os.listdir(folder_full_name)
    i = 1
    for file_name in file_names:
        file_full_name = os.path.join(folder_full_name, file_name)
        folder_name = folder_full_name.split("\\",8)[-1]
        os.rename(file_full_name, folder_full_name + "/" + folder_name + "_" + str(i) + ".jpg")
        i = i + 1          
```

### 5. Rename the test set pictures and create solution.csv

```python
test_folder = dest_folder + "\Test"
```

```python
# create two arrays: picture id and solution
id_array = []
category_array = []

file_names = os.listdir(test_folder)
random.shuffle(file_names)
i = 1
for file_name in file_names:
    file_full_name = os.path.join(test_folder, file_name)
    id_array.append(i)
    category = file_name.split("_")[1]
    category_array.append(category)
    os.rename(file_full_name, test_folder + "\\" + "Test_" + str(i) + ".jpg")
        
    i = i + 1  
```

```python
category_array
```

```python
category_numeric_array = []
for category in category_array:
    if category == 'BabyHat':
        category_numeric_array.append(0)
    if category == 'BabyPants':
        category_numeric_array.append(1)
    if category == 'BabyShirt':
        category_numeric_array.append(2)
    if category == 'PackageFart':
        category_numeric_array.append(3)
    
```

```python
category_numeric_array
```

```python
d = {'id': id_array, 'category': category_numeric_array}
```

```python
df = pd.DataFrame(data=d)
```

```python
cols = ['id', 'category']
df = df[cols]
```

```python
df
```

```python
df.to_csv(dest_folder + "/solution.csv", sep='\t' , index=False, encoding='utf-16')
```
